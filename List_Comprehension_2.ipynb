1) Find all of the numbers from 1-1000 that are divisible by 8.

```python
div8 = [n for n in range(1,1000) if n % 8 == 0] 
print(div8)
```

2) Find all of the numbers from 1-1000 that have a 6 in them. 

```python
six = [n for n in range(0,1000) if '6' in str(n)]
print(six)
```

3) Count the number of spaces in the following string.

```python
def count_spaces(string):
    return  len([char for char in string if char == " "])

string = "My name is Swati Bhogesh"
print("number of spaces ",count_spaces(string))
```

4) Remove all the vowels in the above string.

```python
def remove_vowels(string):
    return [char for char in string if char not in "aeiouy"]

string = "My name is Swati Bhogesh"
print(remove_vowels(string))
```

5) Find all the words in the aboveÂ string that are less than 5 letters.

```python
def find_shorter_than5(string):
    return [word for word in string.split() if len(word) < 5]

string = "My name is Swati Bhogesh"
print(find_shorter_than5(string))
```

1. use dictionary comprehension to count the length of each word in  the sentence 

```python
def length_of_words(string):
    return {word:len(word) for word in string.split()}

string = "My name is Swati Bhogesh"
print(length_of_words(string))
```

2. Use  a nested list comprehension to find all of the numbers from 1-1000 that are divisible by any single digit besides 1(2-9).  

```python
divisible = [num for num in range(1,1001) if [div for div in range(2,10) if num % div == 0]]
print(divisible)
```

3. For all the numbers 1-1000, use a nested list/dictionary comprehension to find the highest single digit any of the numbers is divisible by. 

```python
highest_divisor = {num:max(div for div in range(2,10) if num % div == 0) for num in range(1,1001) if [div for div in range(2,10) if num % div == 0]}
print(highest_divisor)
```
