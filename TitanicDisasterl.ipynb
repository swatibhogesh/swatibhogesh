```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline

# %matplotlib inline : the output of plotting commands is displayed inline within frontends directly below the code cell that produced it. The resulting plots will then also be stored in the notebook document
```

```python
#Step:2 Importing and assigning to a variable
url = 'https://raw.githubusercontent.com/a-forty-two/DFE6/main/titanic.csv'

titanic = pd.read_csv(url)
```

```python
#step 4 
#create a piechart presenting the male/female proportion

#sum of the instances of males and females

males =(titanic['Sex']=='male').sum()
females = (titanic['Sex']=='female').sum()
```

```python
#put them into the list called proportions

proportions =[males, females]
```

```python
proportions
```

```python
#create a piechart
plt.pie(
    #using proportions
    proportions,

    #with the lables 
    labels=["Males",'Females'],

    #with  no shadows
    shadow = False,

    #with colors
    colors =['blue','red'],

    #with one slide exploded out
    explode = (0.15,0),

    #with start angle at 90
    startangle=90,

    #with the percent listed as a fraction

    autopct ="%1.1f%%"
    )
```

```python
plt.axis('equal')

plt.title("Sex proportion")


#View the plot
plt.show()
```

```python
#Creating  a scatter plot with the fare paid and the Age, differ the plot color by gender
lm= sns.lmplot(x="Age",y="Fare", data= titanic,hue="Sex" , fit_reg=0)

lm.set(title= 'Fare X Age')

axes=lm.axes
axes[0,0].set_ylim(-5,)
axes[0,0].set_xlim(-5,85)
```

```python
#Create a histogram with the fare paid
df= titanic.Fare.sort_values(ascending=False)
# Create bins interval using numpy

binsVal= np.arange(0,600,10)
binsVal
```

```python
#create a plot
plt.hist(df,bins= binsVal)

# set the title and labels
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Fare paid Histogram')

plt.show()
```
