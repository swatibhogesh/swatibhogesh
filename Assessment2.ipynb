To create a linked list with firstname,lastname,hobby and phoneno.
1.Add new element to the list. 2.Search and Rename an element from the list 3.Perform deduplication-means all duplicates should be removed 4.Reverse the list

```python
class Node:
  
    
    def __init__(self, firstname , lastname , hobby , ph):
        self.firstname = firstname
        self.lastname = lastname
        self.hobby = hobby
        self.ph = ph
        self.next = None
  
  

class LinkedList:
  
    
    def __init__(self):
        self.head = None


    def insert_node(self, firstname , lastname , hobby , ph):
        node = Node(firstname , lastname , hobby , ph)
        node.next = self.head
        self.head = node
  
  

    def delete_duplicates(self):

        current = self.head
        prev = None
        duplicate_dict = dict()
        while current:
            if current.firstname and current.lastname and current.hobby and current.ph not in duplicate_dict:
                duplicate_dict[current.ph] = None
                prev = current
            else:
                prev.next = current.next

            current = current.next

    
    def printList(self):
        data = self.head
        while (data):
            print(data.firstname , data.lastname, data.hobby , data.ph)
            data = data.next



    def reverse(self):
        prev = None
        current = self.head
        while(current is not None):
            next = current.next
            current.next = prev
            prev = current
            current = next
        self.head = prev



    def Search(self, key):
        current = self.head
 
        index = 0
        while current:
            if current.ph == key:
                current.firstname = "Swati" 
            current = current.next
            index = index + 1
 
        return -1

   
 
  


l = LinkedList()



l.head = Node("Adi" , "Sharma" , "football" , 5562882296)
second = Node("Bhogesh" , "J" , "reading" , 9876437883)
third = Node("Teju" , "Nagendra" , "singing" , 9864672882)
fourth = Node("Megha" , "Jain" , "dancing" , 9876543212)


l.head.next = second  
second.next = third  
third.next = fourth


print("initial LL")
print("\n")
l.printList()

l.insert_node("Megha" , "Jain" , "dancing" , 9876543212)







print("\n")
print("\n")
print("inserting in LL")
print("\n")

l.printList()

l.delete_duplicates()

print("\n")
print("removing duplicates")

print("\n")




l.printList()

l.reverse()

print("\n")

print("reversing")
print("\n")

l.printList()


print("\n")
print("searching")
print("\n")

l.Search(5562882296)

print("\n")

l.printList()
```
